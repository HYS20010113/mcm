```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('Problem_C_Data_Wordle.xlsx')
data
hard=data['Unnamed: 5'][1:]
total=data['Unnamed: 4'][1:]
total=np.array(total)
hard=np.array(hard)
percentageList=[]
for i in range(len(total)):
    percentageList.append(hard[i]/total[i])
percentageList=np.array(percentageList)
np.argmax(percentageList)
```

```python
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
fig,ax=plt.subplots()
months=mdates.MonthLocator()
dateFmt=mdates.DateFormatter("%y/%m/%d")
ax.xaxis.set_major_formatter(dateFmt)
date1=datetime.date(2022,1,7)
date2=datetime.date(2023,1,1)
delta=datetime.timedelta(days=1)
dates=mdates.drange(date1,date2,delta)
percentageList_inv=percentageList[::-1]
ax.plot_date(dates,percentageList_inv,"#FF8800",alpha=0.7)
fig.autofmt_xdate()
plt.show()



```
